```python
import numpy as np
import matplotlib.pyplot as plt

# Завдання 1
# Побудувати графік функції \( y = \exp(2 \cdot \cos(x)) \) на відрізку [0, \( 4\pi \)] з кроком \( \frac{\pi}{15} \) синього кольору з маркерами типу '∙'.

x = np.arange(0, 4 * np.pi, np.pi / 15)
y = np.exp(2 * np.cos(x))

plt.plot(x, y, 'bo-') # Синій графік з маркерами

# Оформлення
plt.title('Графік функції y=exp(2*cos(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Завдання 2
# Додати до попереднього графіку коло зеленого кольору з маркером '*'
x = np.arange(0, 4 * np.pi, np.pi / 15)
y = np.exp(2 * np.cos(x))

fig, ax = plt.subplots()

# Побудова графіка із завдання 1
ax.plot(x, y, 'bo-')  # Синій графік з маркерами

# Додавання кола радіусом 3 і з центром у точці (6,10)
circle = plt.Circle((6, 10), 3, color='green', fill=False, linewidth=2)
ax.add_patch(circle)

# Оформлення
plt.title('Графік функції y=exp(2*cos(x)) з доданим колом')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
ax.set_aspect('equal')  # Щоб круг залишався кругом на графіку
plt.show()
```

```python
# Завдання 3
# Оформлення графіка із Завдання 1 аналогічно прикладу з методичних вказівок
plt.plot(x, y, 'bo-') # Синій графік з маркерами
plt.axis([0, 4*np.pi, 0, np.max(y) + 1])
# Оформлення
plt.title('Графік функції y=exp(2*cos(x))')
plt.xlabel(r'$0 \leq x \leq 4\pi$')
plt.ylabel(r'$y = exp(2\cos(x))$')
plt.text(2*np.pi, np.mean(2*np.pi), r'Пік функції при x = $2\pi$', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Завдання 4
# Побудувати графік функції \( z = x \cdot \exp(-x^2 - y^2) \) в області [-2, 2] з кроком сітки 0.1.
x = np.arange(-2, 2, 0.1)
y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(x, y)
Z = X * np.exp(-X**2 - Y**2)

plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title('Графік функції z = x * exp(-x^2 - y^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# Завдання 5
# Дослідити можливості функцій `surface` та `plot3` для побудови графіку функції з Завдання 4.
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D графік функції z = x * exp(-x^2 - y^2)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
